```python
%pylab notebook
from math import factorial
```

This program calculates a distribution of probablities of quanta of energy
of two blocks of different sizes in thermal equilibrium. It generates a table which ultimately shows how probable a configuration is.

```python
q   = 100     # Total quanta of energy in both blocks
n1  = 300     # Atoms in block 1
n2  = 200     # Atoms in block 2
q1  = 0       # Quanta of energy in block 1
q2  = q - q1  # Quanta of energy in block 2
```

```python
total_states = [] # This empty list is used for generating a histogram.
q_values     = []

print("q1     q2       Omega1      Omega2    Omega1Omega2")
print("====="*10)
while q2 >= 0:
    omega1 = factorial(q1 + n1-1)/(factorial(q1) * factorial(n1-1))
    omega2 = factorial(q2 + n2-1)/(factorial(q2) * factorial(n2-1))
    print("%3d    %3d    %2.2E     %2.2E      %2.2E"
          %(q1, q2, omega1, omega2, omega1*omega2))
    total_states.append(omega1*omega2)
    q_values.append(q1)
    q1+=1
    q2 = q - q1
    
```

```python
plt.scatter(q_values, total_states, alpha = 1)
```
